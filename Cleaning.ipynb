```python
import pandas as pd
import numpy as np
import datetime

clients_id = pd.read_csv("./infos/customers.csv")
products_id = pd.read_csv("./infos/products.csv")
transactions = pd.read_csv("./infos/transactions.csv")
```

```python
# pour clean le fichier products
products_id_c = products_id.loc[products_id["price"]>0]
```

```python
# pour clean le fichier transaction
transactions_c = transactions
transactions_c = transactions_c[~transactions_c["id_prod"].str.contains("T_0")]
transactions_c = transactions_c.reset_index(drop=True)
```
